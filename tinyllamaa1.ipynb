```python
# Install everything
!pip install transformers bitsandbytes peft torch cvxpy numpy matplotlib tabulate accelerate

# Imports
import torch
import numpy as np
import matplotlib.pyplot as plt
import cvxpy as cp
from torch.utils.data import Dataset, DataLoader
from torch.optim import AdamW
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training, TaskType
import re
from tabulate import tabulate
import time
import os # Added for saving model/tokenizer

torch.cuda.empty_cache()

# 1) Convex Oracle (No changes needed)
def generate_optimal_schedule(price, P_h, E_h, peak_limit):
    N, H = len(price), len(P_h)
    X = cp.Variable((N, H))
    power = cp.multiply(X, P_h.reshape(1, H))
    cost = cp.sum(cp.multiply(price.reshape(N, 1), power)) / 1000.0
    constraints = [
        cp.sum(power, axis=0) >= E_h,
        cp.sum(power, axis=1) <= peak_limit,
        X >= 0, X <= 1
    ]
    prob = cp.Problem(cp.Minimize(cost), constraints)
    # Using a robust solver like SCS can sometimes help if default fails
    try:
      prob.solve(verbose=False, solver=cp.SCS)
      if X.value is None: # Fallback if SCS fails or returns None
          print("Warning: SCS solver failed or returned None, trying default solver.")
          prob.solve(verbose=False)
    except cp.SolverError:
        print("Warning: Default solver failed, trying SCS.")
        prob.solve(verbose=False, solver=cp.SCS)

    if X.value is None:
        print("Error: Optimization failed to find a solution. Returning zeros.")
        return np.zeros((N,H), dtype=float) # Return feasible zeros if solver fails

    return X.value.astype(float)

# 2) Dataset (No changes needed)
class TSLSupervisedDatasetStructured(Dataset):
    def __init__(self, n_samples, N, H):
        self.prompts = []
        self.targets = []
        self.N, self.H = N, H
        # Consistent appliance definitions across dataset
        self.P_h = np.array([1000, 1500]) # Watts
        self.E_h = np.array([3000, 6000]) # Watt-hours
        self.peak_range = (2000, 6000) # Watts

        generated_count = 0
        attempts = 0
        max_attempts = n_samples * 2 # Allow for some failures

        print(f"Attempting to generate {n_samples} valid dataset samples...")
        while generated_count < n_samples and attempts < max_attempts:
            attempts += 1
            price = 0.05 + (0.2 - 0.05) * np.random.rand(N) # $/kWh
            peak_limit = np.random.randint(*self.peak_range) # Watts

            # Check basic feasibility: Is total energy required schedulable within peak?
            # Min time needed for Appliance 1 = E_h[0]/P_h[0], for App 2 = E_h[1]/P_h[1]
            # Total energy = sum(E_h)
            # Max energy possible = N * peak_limit
            # This simple check isn't perfect but helps filter impossible scenarios
            if np.sum(self.E_h) > N * peak_limit:
                 # print(f"Skipping infeasible scenario (Peak: {peak_limit}, Energy: {np.sum(self.E_h)})")
                 continue # Skip if peak is clearly too low for total energy over N slots

            X_opt = generate_optimal_schedule(price, self.P_h, self.E_h, peak_limit)

            # Check if the solver actually found a valid solution
            if X_opt is not None and not np.isnan(X_opt).any():
                 # Further check if constraints are reasonably met (allowing for numerical tolerance)
                 power_opt = X_opt * self.P_h.reshape(1, H)
                 energy_met = np.sum(power_opt, axis=0) >= (self.E_h - 1e-3) # Tolerance for float errors
                 peak_ok = np.all(np.sum(power_opt, axis=1) <= (peak_limit + 1e-3)) # Tolerance

                 if np.all(energy_met) and peak_ok:
                    self.prompts.append(self._build_prompt(price, peak_limit))
                    self.targets.append(torch.tensor(X_opt.flatten(), dtype=torch.float32))
                    generated_count += 1
                 # else:
                    # print(f"Skipping sample - constraints not met after solve. Energy met: {energy_met}, Peak ok: {peak_ok}")
            # else:
               # print("Skipping sample - solver failed.")

            if attempts % 100 == 0:
                 print(f"Attempt {attempts}/{max_attempts}, Generated {generated_count}/{n_samples}")


        if generated_count < n_samples:
            print(f"Warning: Could only generate {generated_count}/{n_samples} valid samples.")

        print(f"Successfully generated {generated_count} dataset samples.")


    def _build_prompt(self, price, peak_limit):
        # Use exactly the same parameters as the dataset generation
        s = "<Instruction>\n" # Added structure token
        s += "Given the electricity prices per slot, appliance specifications, and a peak power limit, determine the optimal activation schedule (proportion of time each appliance runs in each slot) to minimize total electricity cost, while ensuring each appliance meets its total energy requirement and the peak power limit is never exceeded.\n\n"
        s += "[Prices]\n" + "\n".join(f"Slot {i+1}: {p:.3f} $/kWh" for i, p in enumerate(price))
        s += "\n\n[Appliances]\n"
        for i, (pw, eg) in enumerate(zip(self.P_h, self.E_h), start=1):
            s += f"Appliance {i}:\n  Rated Power: {pw/1000:.1f} kW\n  Energy Required: {eg/1000:.1f} kWh\n"
        s += f"\n[Peak Limit]\n{peak_limit/1000:.1f} kW\n"
        s += "\n[Objective]\nMinimize total electricity cost.\n"
        # More explicit output format instruction
        s += f"\n[Output Format]\nPlease provide only the activation schedule as a {self.N}x{self.H} matrix.\n"
        s += f"Format it as {self.N} rows, each row containing {self.H} numbers (between 0.0 and 1.0, representing activation proportion) separated by a comma.\n"
        s += "Separate the rows with a semicolon.\n"
        s += "Example row: 0.5,1.0;\n"
        s += "Do not include any other text, explanations, or labels in the output.\n"
        s += "</Instruction>\n<Response>\n" # Added structure token
        return s

    def __len__(self):
        return len(self.prompts)

    def __getitem__(self, idx):
        return self.prompts[idx], self.targets[idx]

# 3) TinyLlama Model with LoRA
class TinyLlamaScheduler:
    def __init__(self, model_name="TinyLlama/TinyLlama-1.1B-Chat-v1.0"): # Changed model name
        self.model_name = model_name
        self.tokenizer = AutoTokenizer.from_pretrained(self.model_name)
        # TinyLlama tokenizer might not have a pad token, set it to eos_token
        if self.tokenizer.pad_token is None:
            self.tokenizer.pad_token = self.tokenizer.eos_token
            print("Set pad_token to eos_token")

        # Quantization configuration
        bnb_config = BitsAndBytesConfig(
            load_in_8bit=True # Use 8-bit quantization
            )

        # Load the model
        model = AutoModelForCausalLM.from_pretrained(
            self.model_name,
            device_map="auto", # Automatically distribute across available GPUs/CPU
            quantization_config=bnb_config,
            # trust_remote_code=True # Often not needed for TinyLlama, try without first
        )

        # Prepare model for k-bit training (important for quantized models)
        model = prepare_model_for_kbit_training(model)

        # LoRA configuration
        lora_config = LoraConfig(
            r=16, # Rank of LoRA matrices (increased from 8)
            lora_alpha=32, # Alpha scaling (increased from 16)
            # Target modules common in Llama models - verify if needed for specific TinyLlama variant
            target_modules=["q_proj", "k_proj", "v_proj", "o_proj", "gate_proj", "up_proj", "down_proj"],
            lora_dropout=0.05,
            bias="none",
            task_type=TaskType.CAUSAL_LM,
        )

        # Apply LoRA adapter
        model = get_peft_model(model, lora_config)
        model.print_trainable_parameters() # Show number of trainable parameters

        self.model = model
        self.model.eval() # Set to evaluation mode initially
        self.N = None # To be set based on dataset
        self.H = None # To be set based on dataset

    def fine_tune(self, dataset, epochs=1, batch_size=1, lr=2e-4):
        self.model.train() # Set model to training mode
        # Use dataset's N and H
        self.N = dataset.N
        self.H = dataset.H
        if self.N is None or self.H is None:
             raise ValueError("Dataset N and H must be set before fine-tuning")

        loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
        optimizer = AdamW(self.model.parameters(), lr=lr)
        loss_history = []

        print(f"Starting fine-tuning with N={self.N}, H={self.H}")

        for ep in range(1, epochs+1):
            running_loss = 0.0
            processed_batches = 0
            for prompts, targets in loader:
                # Construct the full text including prompt and target schedule
                full_texts = []
                for prompt, target_flat in zip(prompts, targets):
                    # Reshape target and format it as text
                    target_matrix = target_flat.reshape(self.N, self.H)
                    # Format with 1 decimal place, comma between cols, semicolon between rows
                    schedule_text = ";".join(",".join(f"{v:.1f}" for v in target_matrix[i]) for i in range(self.N)) + ";" # Ensure trailing semicolon

                    # Append the target schedule to the prompt
                    # Ensure the prompt ends correctly before adding the target response part
                    if prompt.strip().endswith("<Response>"):
                         full_text = prompt.strip() + "\n" + schedule_text + self.tokenizer.eos_token # Add EOS token
                    else: # Add response tag if missing
                         full_text = prompt + "\n<Response>\n" + schedule_text + self.tokenizer.eos_token # Add EOS token

                    full_texts.append(full_text)

                # Tokenize the full texts (prompt + target)
                # Pad to the longest sequence in the batch
                encodings = self.tokenizer(full_texts, return_tensors="pt", padding=True, truncation=True, max_length=1024).to(self.model.device)

                # Standard Causal LM training: labels are the input_ids shifted
                labels = encodings["input_ids"].clone()
                # Optional: Ignore padding tokens in loss calculation
                labels[labels == self.tokenizer.pad_token_id] = -100
                # Optional: Ignore prompt tokens in loss (might improve generation focus)
                # Find where the response starts (look for <Response>\n pattern)
                # This part is tricky and depends heavily on exact tokenization.
                # Simpler approach: just train on the whole sequence for now.

                # Forward pass
                outputs = self.model(**encodings, labels=labels)
                loss = outputs.loss

                # Backward pass and optimization
                optimizer.zero_grad()
                loss.backward()
                optimizer.step()

                running_loss += loss.item()
                processed_batches += 1

            # Calculate average loss for the epoch
            avg_loss = running_loss / processed_batches if processed_batches > 0 else 0.0
            loss_history.append(avg_loss)
            print(f"Fine-tuning Epoch {ep}/{epochs} — Avg Loss: {avg_loss:.6f}")

        self.model.eval() # Set back to evaluation mode

        # Plot Loss Curve
        plt.figure(figsize=(8, 4)) # Adjusted figure size
        plt.plot(range(1, epochs+1), loss_history, marker='o')
        plt.title('Fine-tuning Loss Curve (TinyLlama)')
        plt.xlabel('Epoch')
        plt.ylabel('Average Loss')
        plt.grid(True)
        plt.tight_layout()
        plt.show()

        return self

    def generate_schedule(self, prompt):
        self.model.eval() # Ensure model is in eval mode
        # Make sure the prompt ends with the trigger for the response
        if not prompt.strip().endswith("<Response>"):
             prompt = prompt.strip() + "\n<Response>\n"

        inputs = self.tokenizer(prompt, return_tensors="pt", padding=True, truncation=True, max_length=512).to(self.model.device)

        # Generate output - adjust parameters as needed
        # `do_sample=False` for deterministic output (greedy search)
        # `max_new_tokens` needs to be sufficient for the N x H matrix string
        # Estimate: N rows * (H * ~4 chars/num + H-1 commas + 1 semicolon) + some buffer
        estimated_tokens = self.N * (self.H * 4 + self.H) + 50 # Generous buffer
        with torch.no_grad(): # Ensure no gradients are calculated
            outputs = self.model.generate(
                 **inputs,
                 max_new_tokens=min(estimated_tokens, 400), # Limit max tokens
                 do_sample=False,
                 pad_token_id=self.tokenizer.eos_token_id # Use EOS for padding during generation
                 )

        # Decode only the newly generated tokens
        generated_ids = outputs[0, inputs['input_ids'].shape[1]:]
        generated_text = self.tokenizer.decode(generated_ids, skip_special_tokens=True)

        # Extract the relevant part after the prompt
        # response_tag = "<Response>\n"
        # if response_tag in generated_text_full:
        #     generated_text = generated_text_full.split(response_tag, 1)[1]
        # else:
        #     # Fallback if the tag isn't found as expected (might happen)
        #     # Try to get text after the original prompt length (less reliable)
        #     print("Warning: <Response> tag not found in generation, attempting fallback extraction.")
        #     original_prompt_text = self.tokenizer.decode(inputs['input_ids'][0], skip_special_tokens=True)
        #     if generated_text_full.startswith(original_prompt_text):
        #          generated_text = generated_text_full[len(original_prompt_text):]
        #     else:
        #          generated_text = generated_text_full # Best guess

        return generated_text.strip() # Return only the cleaned generated part

# 4) Parse output (Renamed for generality)
# Regex needs to be robust to extra spaces or slight variations
# It looks for N digits.1digit, H digits.1digit; repeated N times.
def parse_llm_output(output, N, H):
    # Try to find the matrix part specifically, ignoring surrounding text
    # Look for patterns like "0.5,1.0;" or " 0.5 , 1.0 ; "
    # Updated regex: matches digits.digit, allows spaces, expects comma and semicolon
    pattern = r"([\s]*[\d]\.[\d]+[\s]*,[\s]*[\d]\.[\d]+[\s]*;)"
    try:
        # Find all matches for rows
        matches = re.findall(pattern, output)

        if not matches or len(matches) < N:
            print(f"Parsing warning: Found only {len(matches)}/{N} rows matching pattern in output:\n---\n{output}\n---")
            # If few rows found, maybe it stopped early? Pad with zeros.
            # If no rows found, return all zeros.
            schedule = np.zeros((N, H))
            parsed_rows = 0
            for i, row_str in enumerate(matches[:N]): # Use up to N found rows
                 values_str = row_str.strip().strip(';') # Remove outer space/semicolon
                 values = [float(v.strip()) for v in values_str.split(',')]
                 if len(values) == H:
                     schedule[i] = values
                     parsed_rows += 1
                 else:
                     print(f"Parsing warning: Row {i+1} has incorrect number of values ({len(values)}/{H}). String: '{row_str}'")
            print(f"Parsed {parsed_rows} valid rows successfully.")
            if parsed_rows == 0:
                print("Parsing error: Could not extract any valid schedule rows. Using zeros.")
                return np.zeros((N, H))
            # Return partially filled matrix if some rows were parsed
            return np.clip(schedule, 0, 1)

        # If we found at least N rows
        schedule = np.zeros((N, H))
        parsed_rows = 0
        for i, row_str in enumerate(matches[:N]): # Take the first N matches
            values_str = row_str.strip().strip(';') # Remove outer space/semicolon
            try:
                values = [float(v.strip()) for v in values_str.split(',')]
                if len(values) == H:
                    schedule[i] = values
                    parsed_rows += 1
                else:
                    print(f"Parsing warning: Row {i+1} has incorrect number of values ({len(values)}/{H}). String: '{row_str}'")
                    # Optionally leave this row as zeros or handle differently
            except ValueError as e:
                print(f"Parsing error in row {i+1}: Could not convert values. String: '{row_str}'. Error: {e}")
                # Optionally leave this row as zeros

        print(f"Parsed {parsed_rows}/{N} rows successfully.")
        # Clip final result to ensure bounds [0, 1]
        return np.clip(schedule, 0, 1)

    except Exception as e:
        print(f"Critical Parsing error: {e}. Output was:\n---\n{output}\n---")
        print("Using zeros as fallback.")
        return np.zeros((N, H))


# 5) Test and Plot Function (Updated to use generic parser, removed use_cache)
def test_and_plot(scheduler_instance, tests=5, N=25, H=2):
    import matplotlib.pyplot as plt
    import numpy as np

    # Use parameters consistent with the dataset generation
    P_h = np.array([1000, 1500])
    E_h = np.array([3000, 6000])
    peak_range = (2000, 6000)

    # Get components from the scheduler instance
    tokenizer = scheduler_instance.tokenizer
    model = scheduler_instance.model # PEFT model
    model.eval() # Ensure eval mode

    results = []

    print(f"\n--- Running {tests} Test Scenarios ---")

    for t_i in range(1, tests + 1):
        print(f"\n--- Test Scenario {t_i}/{tests} ---")
        # Generate a new test case
        price = 0.05 + (0.2 - 0.05) * np.random.rand(N)
        peak = np.random.randint(*peak_range)

        # Build Prompt using the same logic as dataset generation
        # Need the prompt builder logic from the dataset class or replicate it here
        # Replicating builder logic for consistency:
        def _build_test_prompt(price, peak_limit, P_h, E_h, N, H):
             s = "<Instruction>\n" # Added structure token
             s += "Given the electricity prices per slot, appliance specifications, and a peak power limit, determine the optimal activation schedule (proportion of time each appliance runs in each slot) to minimize total electricity cost, while ensuring each appliance meets its total energy requirement and the peak power limit is never exceeded.\n\n"
             s += "[Prices]\n" + "\n".join(f"Slot {i+1}: {p:.3f} $/kWh" for i, p in enumerate(price))
             s += "\n\n[Appliances]\n"
             for i, (pw, eg) in enumerate(zip(P_h, E_h), start=1):
                 s += f"Appliance {i}:\n  Rated Power: {pw/1000:.1f} kW\n  Energy Required: {eg/1000:.1f} kWh\n"
             s += f"\n[Peak Limit]\n{peak_limit/1000:.1f} kW\n"
             s += "\n[Objective]\nMinimize total electricity cost.\n"
             s += f"\n[Output Format]\nPlease provide only the activation schedule as a {N}x{H} matrix.\n"
             s += f"Format it as {N} rows, each row containing {H} numbers (between 0.0 and 1.0, representing activation proportion) separated by a comma.\n"
             s += "Separate the rows with a semicolon.\n"
             s += "Example row: 0.5,1.0;\n"
             s += "Do not include any other text, explanations, or labels in the output.\n"
             s += "</Instruction>\n<Response>\n" # Added structure token
             return s

        prompt_text = _build_test_prompt(price, peak, P_h, E_h, N, H)
        # print(f"Test Prompt:\n{prompt_text}") # Optional: print the prompt

        # Generate schedule using the model's method
        generated_output = scheduler_instance.generate_schedule(prompt_text)
        print(f"Generated Output (Raw):\n---\n{generated_output}\n---")

        # Parse prediction using the updated function
        pred_sched = parse_llm_output(generated_output, N, H)
        print(f"Parsed Schedule (Predicted):\n{tabulate(pred_sched, floatfmt='.2f')}")

        # Solve for the optimal schedule using the oracle
        opt_sched = generate_optimal_schedule(price, P_h, E_h, peak)
        if opt_sched is None or np.isnan(opt_sched).any():
             print("Warning: Optimal solver failed for this test case. Skipping comparison plots/metrics.")
             continue # Skip to next test if optimum not found

        print(f"Optimal Schedule:\n{tabulate(opt_sched, floatfmt='.2f')}")


        # --- Validation ---
        # Check constraints for Predicted Schedule
        pred_power = pred_sched * P_h.reshape(1, H)
        pred_energy_total = np.sum(pred_power, axis=0)
        pred_peak_usage = np.sum(pred_power, axis=1)
        pred_energy_met = pred_energy_total >= (E_h - 1) # Allow small tolerance
        pred_peak_ok = np.all(pred_peak_usage <= (peak + 1)) # Allow small tolerance
        pred_valid = np.all(pred_energy_met) and pred_peak_ok

        # Check constraints for Optimal Schedule (should always be valid if solver worked)
        opt_power = opt_sched * P_h.reshape(1, H)
        opt_energy_total = np.sum(opt_power, axis=0)
        opt_peak_usage = np.sum(opt_power, axis=1)
        opt_energy_met = opt_energy_total >= (E_h - 1)
        opt_peak_ok = np.all(opt_peak_usage <= (peak + 1))
        opt_valid = np.all(opt_energy_met) and opt_peak_ok


        # Compute Costs
        cost_pred = np.sum(price.reshape(N, 1) * pred_power) / 1000.0
        cost_opt = np.sum(price.reshape(N, 1) * opt_power) / 1000.0 if opt_valid else float('inf') # Use inf if opt is invalid


        gap = cost_pred - cost_opt if opt_valid else float('nan')
        optimality_ratio = (cost_pred / cost_opt) * 100 if cost_opt > 1e-6 and opt_valid else float('inf') # Avoid division by zero

        results.append({
            "Test": t_i,
            "Cost (Pred)": cost_pred,
            "Cost (Opt)": cost_opt,
            "Gap ($)": gap,
            "Optimality (%)": optimality_ratio,
            "Pred Valid": "Yes" if pred_valid else f"No (E:{pred_energy_met} P:{pred_peak_ok})",
            "Opt Valid": "Yes" if opt_valid else "No", # Should ideally always be Yes
        })
        print(f"Prediction valid: {pred_valid}, Optimal valid: {opt_valid}")
        print(f"Energy Pred: {pred_energy_total.round(1)} (Req: {E_h})")
        print(f"Max Peak Pred: {np.max(pred_peak_usage).round(1)} (Limit: {peak})")
        print(f"Energy Opt: {opt_energy_total.round(1)} (Req: {E_h})")
        print(f"Max Peak Opt: {np.max(opt_peak_usage).round(1)} (Limit: {peak})")
        print("-" * 20)

        # Plot
        t = np.arange(N)
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8), sharex=True) # Wider figure

        # Predicted Plot
        ax1.bar(t, pred_power[:,0], label=f"Appliance 1 (E={pred_energy_total[0]:.0f}/{E_h[0]})")
        ax1.bar(t, pred_power[:,1], bottom=pred_power[:,0], label=f"Appliance 2 (E={pred_energy_total[1]:.0f}/{E_h[1]})")
        ax1.plot(t, pred_peak_usage, 'g--', label=f'Total Usage (Max: {np.max(pred_peak_usage):.0f}W)')
        ax1.axhline(peak, color='r', linestyle=':', linewidth=2, label=f"Peak Limit ({peak}W)")
        ax1.set_title(f"Fine-Tuned TinyLlama Prediction (Test {t_i}) - Cost: ${cost_pred:.3f}" + (" (INVALID)" if not pred_valid else ""))
        ax1.set_ylabel("Power (W)")
        ax1.legend(loc='upper right')
        ax1.grid(True, axis='y', linestyle='--', alpha=0.7)
        ax1.set_ylim(bottom=0) # Ensure y-axis starts at 0

        # Optimal Plot
        ax2.bar(t, opt_power[:,0], label=f"Appliance 1 (E={opt_energy_total[0]:.0f}/{E_h[0]})")
        ax2.bar(t, opt_power[:,1], bottom=opt_power[:,0], label=f"Appliance 2 (E={opt_energy_total[1]:.0f}/{E_h[1]})")
        ax2.plot(t, opt_peak_usage, 'g--', label=f'Total Usage (Max: {np.max(opt_peak_usage):.0f}W)')
        ax2.axhline(peak, color='r', linestyle=':', linewidth=2, label=f"Peak Limit ({peak}W)")
        ax2.set_title(f"Optimal Convex Schedule (Test {t_i}) - Cost: ${cost_opt:.3f}")
        ax2.set_xlabel("Time Slot")
        ax2.set_ylabel("Power (W)")
        ax2.legend(loc='upper right')
        ax2.grid(True, axis='y', linestyle='--', alpha=0.7)
        ax2.set_ylim(bottom=0) # Ensure y-axis starts at 0

        plt.tight_layout()
        plt.show()

        # print(f"Test {t_i}: Predicted Cost = ${cost_pred:.3f} | Optimal Cost = ${cost_opt:.3f} | Gap = ${gap:.3f} | Optimality = {optimality_ratio:.2f}%")
        # print(f"  Prediction Validity: {pred_valid} | Optimal Validity: {opt_valid}")

    # Print summary table
    print("\n--- Test Summary ---")
    if results:
         headers = results[0].keys()
         rows = [list(r.values()) for r in results]
         # Format floats in rows for better table display
         formatted_rows = []
         for row in rows:
             new_row = []
             for item in row:
                 if isinstance(item, float):
                     new_row.append(f"{item:.3f}" if abs(item) < 1000 else f"{item:.2f}") # Adjust precision based on value
                 else:
                     new_row.append(item)
             formatted_rows.append(new_row)

         print(tabulate(formatted_rows, headers=headers, tablefmt="grid"))

         # Calculate Average Metrics for valid predictions compared to valid optimals
         valid_gaps = [r["Gap ($)"] for r in results if isinstance(r["Gap ($)"], float) and not np.isnan(r["Gap ($)"]) and r["Pred Valid"]=="Yes"]
         valid_ratios = [r["Optimality (%)"] for r in results if isinstance(r["Optimality (%)"], float) and not np.isinf(r["Optimality (%)"]) and r["Pred Valid"]=="Yes"]
         num_pred_valid = sum(1 for r in results if r["Pred Valid"]=="Yes")

         avg_gap = np.mean(valid_gaps) if valid_gaps else float('nan')
         avg_ratio = np.mean(valid_ratios) if valid_ratios else float('nan')
         print(f"\nAverage Gap (Valid Preds): ${avg_gap:.3f}")
         print(f"Average Optimality Ratio (Valid Preds): {avg_ratio:.2f}%")
         print(f"Valid Predictions: {num_pred_valid}/{tests}")
    else:
        print("No results to summarize.")


# 6) Main Execution Block
if __name__ == "__main__":
    start_time = time.perf_counter() # Record the start time

    # --- Configuration ---
    N, H = 25, 2 # Time slots, Appliances
    DATASET_SIZE = 500 # Number of samples to generate for training (adjust based on time/memory)
    EPOCHS = 10 # Number of fine-tuning epochs (adjust based on convergence)
    BATCH_SIZE = 8  # Batch size for training (adjust based on GPU memory - 16 might be too large for 1.1B model in 8bit)
    LEARNING_RATE = 1e-4 # Learning rate (adjust based on training dynamics)
    NUM_TESTS = 5     # Number of test scenarios to evaluate
    MODEL_SAVE_PATH = "./tinyllama-scheduler-model" # Directory to save the fine-tuned model
    TOKENIZER_SAVE_PATH = "./tinyllama-scheduler-tokenizer" # Directory to save the tokenizer

    # --- Dataset Generation ---
    print(f"\n--- Generating Dataset ({DATASET_SIZE} samples) ---")
    ds_start_time = time.perf_counter()
    ds = TSLSupervisedDatasetStructured(DATASET_SIZE, N, H)
    ds_end_time = time.perf_counter()
    if len(ds) == 0:
        print("FATAL: Dataset generation failed. Exiting.")
        exit()
    elif len(ds) < DATASET_SIZE:
        print(f"Warning: Generated only {len(ds)} samples out of requested {DATASET_SIZE}.")
    print(f"Dataset generated in {ds_end_time - ds_start_time:.2f} seconds.")

    # --- Model Loading ---
    print("\n--- Loading TinyLlama with LoRA ---")
    load_start_time = time.perf_counter()
    # Instantiate the scheduler (loads model, applies LoRA)
    scheduler = TinyLlamaScheduler()
    # Set N, H in the scheduler instance based on the dataset
    scheduler.N, scheduler.H = ds.N, ds.H
    load_end_time = time.perf_counter()
    print(f"Model loaded in {load_end_time - load_start_time:.2f} seconds.")

    # --- Fine-tuning ---
    print(f"\n--- Fine-tuning (LoRA) for {EPOCHS} Epochs ---")
    tune_start_time = time.perf_counter()
    scheduler = scheduler.fine_tune(ds, epochs=EPOCHS, batch_size=BATCH_SIZE, lr=LEARNING_RATE)
    tune_end_time = time.perf_counter()
    print(f"Fine-tuning completed in {tune_end_time - tune_start_time:.2f} seconds.")

    # --- Testing ---
    print(f"\n--- Testing Model on {NUM_TESTS} Scenarios ---")
    test_start_time = time.perf_counter()
    test_and_plot(scheduler, tests=NUM_TESTS, N=N, H=H)
    test_end_time = time.perf_counter()
    print(f"Testing completed in {test_end_time - test_start_time:.2f} seconds.")

    # --- Saving Model and Tokenizer ---
    print("\n--- Saving Fine-tuned Model and Tokenizer ---")
    save_start_time = time.perf_counter()
    try:
        os.makedirs(MODEL_SAVE_PATH, exist_ok=True)
        os.makedirs(TOKENIZER_SAVE_PATH, exist_ok=True)

        # Save the PEFT model adapter weights
        scheduler.model.save_pretrained(MODEL_SAVE_PATH)
        print(f"PEFT Model saved to: {MODEL_SAVE_PATH}")

        # Save the tokenizer
        scheduler.tokenizer.save_pretrained(TOKENIZER_SAVE_PATH)
        print(f"Tokenizer saved to: {TOKENIZER_SAVE_PATH}")

    except Exception as e:
        print(f"Error saving model/tokenizer: {e}")
    save_end_time = time.perf_counter()
    print(f"Saving completed in {save_end_time - save_start_time:.2f} seconds.")


    # --- Total Time ---
    end_time = time.perf_counter() # Record the end time
    total_elapsed_time = end_time - start_time # Calculate the difference

    # Format the time for better readability
    minutes = int(total_elapsed_time // 60)
    seconds = total_elapsed_time % 60

    print("\n" + "="*40)
    print(f"Total script execution time: {total_elapsed_time:.2f} seconds")
    if minutes > 0:
        print(f"Which is approximately {minutes} minute(s) and {seconds:.2f} second(s).")
    print("="*40)

```
